### LAKERS

```python
import pandas as pd

lakers = pd.read_html("https://www.espn.com/nba/team/stats/_/name/lal/los-angeles-lakers")
```

```python
lakers_name = lakers[0]
lakers_stats = lakers[1]
```

```python
lakers = pd.concat([lakers_name,lakers_stats],axis=1)
lakers
lakers = lakers.drop(24)
lakers
```

```python
lakers.iloc[:,2:13] = lakers.iloc[:,2:13].astype(int)

lakers.shape
```

```python
lakers = lakers.rename(index = {0:"Los Angeles Lakers"})
```

```python
lakers
```

### MIAMI HEAT

```python
miami = pd.read_html("https://www.espn.com/nba/team/stats/_/name/mia/miami-heat")
```

```python
miami_names = miami[0]
miami_stats = miami[1]

miami = pd.concat([miami_names,miami_stats],axis=1)
miami
```

```python
miami = miami.drop(20)
```

```python
miami[['GS','MIN','PTS','OR','DR','REB','AST','STL','BLK','TO','PF','AST/TO']] = miami[['GS','MIN','PTS','OR','DR','REB','AST','STL','BLK','TO','PF','AST/TO']].astype(int)
```

```python
miami = miami.rename(index ={0:"MIAMI HEAT"})
miami
```

### NEW ORLEANS PELICANS

```python
orleans = pd.read_html("https://www.espn.com/nba/team/stats/_/name/no/new-orleans-pelicans")
```

```python
orleans_name = orleans[0]
orleans_stats = orleans[1]

orleans = pd.concat([orleans_name,orleans_stats], axis = 1)
```

```python
orleans
```

```python
orleans = orleans.drop(17)
```

```python
orleans.iloc[:,2:] = orleans.iloc[:,2:].astype(int)

orleans
```

```python
orleans = orleans.rename(index ={0:'NEW ORLEANS PELICANS'})
orleans
```

```python
merged_data = pd.concat([lakers,miami,orleans], axis = 0, ignore_index = False)
```
