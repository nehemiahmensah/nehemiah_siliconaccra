### LAKERS

```python
import pandas as pd

lakers = pd.read_html("https://www.espn.com/nba/team/stats/_/name/lal/los-angeles-lakers")
```

```python
lakers_name = lakers[0]
lakers_stats = lakers[1]
```

```python
lakers = pd.concat([lakers_name,lakers_stats],axis=1)
lakers
lakers = lakers.drop(24)
lakers['Team'] = 'Los Angeles Lakers'
lakers[['DR','TO']] = lakers[['DR','TO']].astype(int)
```

```python
lakers
```

### MIAMI HEAT

```python
miami = pd.read_html("https://www.espn.com/nba/team/stats/_/name/mia/miami-heat")
```

```python
miami_names = miami[0]
miami_stats = miami[1]

miami = pd.concat([miami_names,miami_stats],axis=1)
miami = miami.drop(20)
miami['Team'] = 'Miami Heat'
miami[['DR','TO']] = miami[['DR','TO']].astype(int)
miami
```

### NEW ORLEANS PELICANS
