#####  Q2: SELECT ALL THE RESTAURANTS WHOSE RATING COUNT IS GREATER THAN 10 AND RATING MEAN GREATER THAN THE AVERAGE RATING MEAN OF ALL THE RESTAURANT

```python
df.rating = df.rating.replace("Not given", np.nan)
```

```python
df = df.astype({'rating':'float'})
```

```python
new = df.groupby('restaurant_name').sum('rating').sort_values(by = 'rating', ascending = False).reset_index()
```

```python
res_rating = df.groupby("restaurant_name").count().sort_values(by = 'rating', ascending = False).reset_index().iloc[:,[0,6]]
```

```python
res_rating.iloc[:,-1]
```

```python
new['rating_count'] = res_rating.iloc[:,-1]
```

```python
new
```

```python
new['res_mean'] = new.rating/new.rating_count 
```

```python
new['all_res_mean'] = new.res_mean.sum()/len(new.rating)
```

```python
new
```

```python
new.groupby('restaurant_name').sum('rating').query("`rating`>10 and `res_mean` > `all_res_mean`").sort_values(by = 'rating',ascending = False)
```
